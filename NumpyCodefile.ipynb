```python
M = 32
N = 64
xmin = ymin = 0.0
xmax = ymax = 1.0

x = np.linspace(xmin, xmax, M, endpoint=False)
y = np.linspace(ymin, ymax, N, endpoint=False)

print(x.shape)
print(y.shape)
y
```

```python
import time
import numpy as np
M = 32
N = 64

x = np.linspace(0.0, 1.0, M, endpoint=False)
y = np.linspace(0.0, 1.0, N, endpoint=False)
```

```python
t0 = time.time()

g = np.zeros((M,N))
print(g)

for i in range(M):
    for j in range(N):
        g[i,j] = np.sin(2.0*np.pi*x[i]*y[j])

print(g)
t1 = time.time()
print("time elapsed: {} s".format(t1-t0))
```

```python
x2d, y2d = np.meshgrid(x, y, indexing="ij")

print(x2d[:,0])
print(x2d[0,:])

print(y2d[:,0])
print(y2d[0,:])

print(x2d)
```

```python
t0 = time.time()
g2 = np.sin(2.0*np.pi*x2d*y2d)
t1 = time.time()
print("time elapsed: {} s".format(t1-t0))
```

Splitting of arrays:

```python
x = [2, 4, 5, 3, 6, 9, 4, 1, 8 , 7]
x1, x2, x3, x4 = np.split(x, [4, 6, 7])
print(x1, x2, x3, x4)
```
